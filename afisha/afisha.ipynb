**Как часто люди возвращаются.**  
Расчёт Retention Rate.

Определим событие и период — на их основе сформируем когорту. 
Возьмём дату, когда пользователь впервые проявил активность.
Эти данные запишем в объект Series first_session. 
Объединим его с исходным датафреймом visits:

```python
first_session = visits.groupby('uid')['start_ts'].min()
first_session.name = 'first_session'
visits = pd.merge(visits,first_session,on='uid')
```

Следующим шагом выделим из столбцов first_session_month и session_month месяцы. 
Вызовем метод astype(): преобразуем значения столбцов к типу 'datetime64[M]'. 

```python
visits['first_month'] = visits['first_session'].astype('datetime64[M]')
visits['session_month'] = visits['session_date'].astype('datetime64[M]')
```

Рассчитаем время жизни когорты.

```python
visits['cohort_lifetime'] = ((visits['session_month'] - visits['first_month']) / np.timedelta64(1,'M')).round().astype('int')
visits['first_month'] = visits['first_month'].dt.strftime('%Y-%m')
```

Когортой станет столбец first_month — месяц, в котором была открыта первая сессия.

```python
cohorts = visits.groupby(['first_month','cohort_lifetime'])['uid'].nunique().reset_index()
```

Найдём исходное количество пользователей в когорте.

```python
inital_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_month','uid']]
inital_users_count = inital_users_count.rename(columns={'uid':'cohort_users'})
inital_users_count.head()
```

Объединим данные по когортам с исходным количеством пользователей в когорте:

```python
cohorts = cohorts.merge(inital_users_count,on='first_month')
```

Рассчитаем Retention Rate. Разделим количество активных пользователей в каждый из месяцев на исходное число пользователей в когорте:

```python
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users']
```

Построим сводную таблицу и создадим тепловую карту:

```python
retention_pivot = cohorts[cohorts['cohort_lifetime'] > 0].pivot_table(
    index='first_month',
    columns='cohort_lifetime',
    values='retention',aggfunc='sum')
retention_pivot.fillna('')

plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
g=sns.heatmap(retention_pivot, annot=True, fmt='.1%',vmax=0.10,cmap="YlGnBu");
g.set_yticklabels(g.get_yticklabels(), rotation=45)
```

Коэффициент удержания чрезвычайно низок. Наиболее оптимистично картина выглядит в сентябре - октябре, это время премьер, возвращения большей части потенциальной аудитории сервиса из отпусков. Нужно работать над удержанием пользователей сервиса в остальное время.

Расчёт Retention Rate в зависимости от устройства.

```python
def Retention_Rate_device_heatmap(df,title):

    cohorts = df.groupby(['first_month','cohort_lifetime'])['uid'].nunique().reset_index()

    inital_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_month','uid']]
    inital_users_count = inital_users_count.rename(columns={'uid':'cohort_users'})
    inital_users_count.head()

    cohorts = cohorts.merge(inital_users_count,on='first_month')

    cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users']

    retention_pivot = cohorts[cohorts['cohort_lifetime'] > 0].pivot_table(
        index='first_month',
        columns='cohort_lifetime',
        values='retention',aggfunc='sum')
    retention_pivot.fillna('')

    plt.figure(figsize=(13, 9))
    plt.title(title)
    sns.heatmap(retention_pivot, annot=True, fmt='.1%',vmax=0.10,cmap="YlGnBu");

```

**Сколько раз покупают за период?**

```python
first_purchase = orders.groupby('uid')['order_date'].min().reset_index()
first_purchase.columns = ['uid', 'first_order']
order_frequency = pd.merge(orders, first_purchase, on='uid')

def time_diff(row):
    date1 = row['order_date']
    date2 = row['first_order']
    return (date1.year - date2.year) * 12 + date1.month - date2.month

order_frequency['time_difference'] = order_frequency.apply(time_diff, axis=1)
order_frequency['first_order'] = order_frequency['first_order'].astype('datetime64[M]').dt.strftime('%Y-%m')
orders_grouped_cohorts = (order_frequency
                                                .groupby(['first_order','time_difference'])
                                                .agg({'order_date': 'count', 'uid': 'nunique'})
                               )

orders_grouped_cohorts['order_per_user'] = orders_grouped_cohorts['order_date']/orders_grouped_cohorts['uid']
orders_grouped_cohorts = orders_grouped_cohorts.reset_index()
orders_pivot = orders_grouped_cohorts.pivot_table(index='first_order', 
                                                  columns='time_difference', 
                                                  values='order_per_user')

orders_per_user = orders_grouped_cohorts.groupby('first_order').agg({'order_per_user':'mean'}).round(2)
orders_per_user.columns = ['Среднее количествово заказов на пользователя в месяц']
orders_per_user
```

**Какой средний чек?**

```python
orders_monthly = orders
orders_monthly['order_date'] = orders_monthly['order_date'].astype('datetime64[M]')

first_orders_month = orders_monthly.groupby('uid').agg({'order_date': 'min'}).reset_index()
first_orders_month.columns = ['uid', 'first_order']

cohort_sizes = first_orders_month.groupby('first_order').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order', 'n_buyers']

orders_first_orders_month = orders.merge(first_orders_month, on='uid')
cohorts = (orders_first_orders_month
                                    .groupby(['order_date', 'first_order'])
                                    .agg({'revenue': 'sum', 'uid': 'count'})
                                    .reset_index()
           )

mean_check = cohorts.merge(cohort_sizes, on='first_order')

time_diff = mean_check['order_date'] - mean_check['first_order']
mean_check['cohort_lifetime'] = time_diff / np.timedelta64(1, 'M')
mean_check['cohort_lifetime'] = mean_check['cohort_lifetime'].round().astype('int')
mean_check['mean_check'] = mean_check['revenue']/mean_check['uid']
mean_check['first_order'] = mean_check['first_order'].astype('datetime64[M]').dt.strftime('%Y-%m')
mean_check_pivot = mean_check.pivot_table(index='first_order', columns='cohort_lifetime', values='mean_check')
plt.figure(figsize=(13, 9))
plt.title('Средний чек на пользователя в месяц')
sns.heatmap(mean_check_pivot, annot=True, fmt='.2f', linewidths=1, cmap='YlGnBu')
```

**Сколько денег приносят? (LTV)**  

```python
#margin_rate = 0.5
margin_rate = 1
LTV = cohorts.merge(cohort_sizes, on='first_order')
LTV['gp'] = LTV['revenue'] * margin_rate
LTV['age'] = (LTV['order_date'] - LTV['first_order']) / np.timedelta64(1, 'M')
LTV['age'] = LTV['age'].round().astype('int')
LTV['ltv'] = LTV['gp'] / LTV['n_buyers']
LTV['first_order'] = LTV['first_order'].astype('datetime64[M]').dt.strftime('%Y-%m')
LTV_pivot = LTV.pivot_table(index="first_order",
                columns="age",
                values="ltv",
                aggfunc='mean')

plt.figure(figsize=(13, 9))
plt.title('LTV')
sns.heatmap(LTV_pivot, annot=True, fmt='.2f', linewidths=1, cmap='YlGnBu')
```

```python
ltv_1 = LTV_pivot.cumsum(axis=1)
plt.figure(figsize=(13, 9))
plt.title('Средний LTV')
sns.heatmap(ltv_1, annot=True, fmt='.2f', linewidths=1, cmap='YlGnBu')
```

**Сколько стоило привлечение одного покупателя из каждого источника?**

Посчитаем CAC. Поделим расходы за месяц на количество людей в когорте

```python
visits_sorted = visits.sort_values(by='start_ts')

costs_per_source = costs.groupby('source_id').agg({'costs': 'sum'})
first_source_per_user = visits_sorted.groupby('uid').agg({'source_id': 'first'})
client_per_source = orders.merge(first_source_per_user, on='uid')
client_per_source = client_per_source.groupby('source_id').agg({'uid': 'nunique'})
uid_per_source = visits_sorted.groupby('source_id').agg({'uid': 'nunique'})

rev_per_source = orders.merge(first_source_per_user, on='uid')
rev_per_source = rev_per_source.groupby('source_id').agg({'revenue': 'sum'})

report_uid_costs_per_source = pd.concat([costs_per_source, rev_per_source, 
                                      uid_per_source, client_per_source],axis=1).fillna(0)
report_uid_costs_per_source.columns = ['costs', 'revenue', 'uid', 'users']
report_uid_costs_per_source['cac'] = report_uid_costs_per_source['costs']/report_uid_costs_per_source['users']
report_uid_costs_per_source['cac'] = report_uid_costs_per_source['cac'].round(2)
report_uid_costs_per_source['cac'].sort_values(ascending=False).to_frame()
```

```python
margin_rate = 0.5
report_uid_costs_per_source['gp'] = report_uid_costs_per_source['revenue'] * margin_rate
report_uid_costs_per_source['ltv'] = report_uid_costs_per_source['gp'] / report_uid_costs_per_source['users']
report_uid_costs_per_source['roi'] = report_uid_costs_per_source['ltv'] / report_uid_costs_per_source['cac']
df = report_uid_costs_per_source.reset_index()
roi_by_source = df.pivot_table(index='source_id', values=['roi', 'ltv', 'cac', 'costs', 'gp', 'users'])
roi_by_source.round(2)
```
